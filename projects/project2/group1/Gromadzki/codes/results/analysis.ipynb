```python
import pandas as pd
import numpy as np
import os
```

```python
# load all the data from this path
files = os.listdir()
dfs = [pd.read_csv(f) for f in files if f.endswith(".csv")]
dfs = dfs[2:]
```

```python
# concat all dfs
df = pd.concat(dfs)
df.sort_values(by="metric", ascending=False)
```
