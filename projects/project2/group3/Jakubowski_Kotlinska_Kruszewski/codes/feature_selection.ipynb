## Feature selection

```python
from feature_selection import get_importance, plot_importance, test_k, plot_k, test_combinations, plot_combinations
import pandas as pd
import numpy as np
from evaluation import evaluate
from sklearn.neural_network import MLPClassifier as MLP
from sklearn.svm import SVC
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
from sklearn.preprocessing import StandardScaler
```

### Load data

```python
X_train = pd.read_csv('../data/x_train.txt', header=None, delim_whitespace=True)
y_train = pd.read_csv('../data/y_train.txt', header=None, delim_whitespace=True)
X_test = pd.read_csv('../data/x_test.txt', header=None, delim_whitespace=True)
y_train = np.array(y_train).ravel()
```

### Get feature importances with Random Forest

```python
top_features, top_std = get_importance(X_train, y_train, 10)
plot_importance(top_features, top_std)
```

### Test models performance across different feature numbers

From our initial exploration the best performing models are MLP, SVM and QDA. Therefore, we will check with how many features these models yield the best results in terms of evaluation function. As penalty for number of features is relatively big, it is reasonable to test k values ranging from 3 to 6.

```python
models = [SVC(probability=True, random_state=42), MLP(random_state=42), QDA()]
k_values = [3, 4, 5, 6]

# scale data
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns=X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns=X_test.columns)

# test k values
results = test_k(X_train, y_train, models, k_values, evaluate)
```

```python
results['model'] = results['model'].map({'SVC': 'SVM', 'MLPClassifier': 'MLP', 'QuadraticDiscriminantAnalysis': 'QDA'})
plot_k(results)
```

We will continue our experiments with k set to 4.

### Test different feature combinations

It turned out from initial experiments that although tree-based models generally select the same features as the most important ones, their ranking can slightly differ, therefore we will test different combinations of 4 out of 5 commonly selected features.

```python
top_features = [100, 101, 102, 103, 105]
k = 4
results = test_combinations(X_train, y_train, models, top_features, k)
```

```python
results['model'] = results['model'].map({'SVC': 'SVM', 'MLPClassifier': 'MLP', 'QuadraticDiscriminantAnalysis': 'QDA'})
plot_combinations(results)
```

It turns out that the best set of features for every model is 101, 102, 103 and 105. Therefore we will utilize only these features in further experiments.

```python
best_features = [101, 102, 103, 105]
X_train_selected = X_train[best_features]
X_test_selected = X_test[best_features]

# save selected features
X_train_selected.to_csv('../data/selected/x_train_selected.txt', header=None, index=None, sep=' ')
X_test_selected.to_csv('../data/selected/x_test_selected.txt', header=None, index=None, sep=' ')
```
